```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('datasets\\mercc_cleaned.csv')

df = data[['user_id','session_id','stime','item_id','event_id']]
```

```python
df = df.sort_values(['user_id','stime'])

df['next_action'] = df.groupby(['user_id','session_id','item_id'])['event_id'].shift(-1)

df = df.dropna(subset=['next_action'])

transition_matrix = pd.crosstab(
    df['event_id'],
    df['next_action'],
    normalize='index'
)
```

```python
print("--- Transition Probability Matrix ---")
print(transition_matrix.round(2))
```

```python
print("--- Transition Probability Matrix (Percentage) ---")
print(transition_matrix*100)
```
